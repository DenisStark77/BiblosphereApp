```python
def crop_and_rotate(img, contour, theta):
    # rotate img
    if theta > np.pi / 2.0:
        theta = theta - np.pi

    rows, cols = img.shape[0], img.shape[1]

    contour[:, :, 0] = contour[:, :, 0] * cols
    contour[:, :, 1] = contour[:, :, 1] * rows
    contour = np.int0(contour)

    x, y, w, h = cv2.boundingRect(contour)
    if x < 0:
        w = w + x
        x = 0

    if y < 0:
        h = h + y
        y = 0

    # crop source
    img_crop = img[y:y + h, x:x + w]
    contour[:, :, 0] = contour[:, :, 0] - x
    contour[:, :, 1] = contour[:, :, 1] - y

    rx = contour[0, 0, 0]
    ry = contour[0, 0, 1]

    # If top left corner out of the range use intersection with left side
    if rx < 0:
        ry = - contour[0, 0, 0]*(contour[0, 2, 1] - contour[0, 0, 1])/(contour[0, 2, 0] - contour[0, 0, 0])
        rx = 0

    M = cv2.getRotationMatrix2D((rx, ry), theta / np.pi * 180, 1)

    # rotate contour first to get target size
    contour_rot = np.int0(cv2.transform(contour, M))[0]
    contour_rot[contour_rot < 0] = 0
    x, y, w, h = cv2.boundingRect(contour_rot)

    img_rot = cv2.warpAffine(img_crop, M, (w+rx, h))

    mask = np.zeros(img_rot.shape[0:2])
    cv2.drawContours(mask, np.array([contour_rot]), -1, 255, -1)
    img_masked = np.zeros_like(img_rot)
    img_masked[mask == 255] = img_rot[mask == 255]

    # crop
    img_crop = img_masked[y:y + h, x:x + w]

    return img_crop


def get_contours(img):
    # Resize to 512 if larger
    if img.shape[1] > 512:
        ratio = 512/img.shape[1]
        img = cv2.resize(img, dsize=(0, 0), fx=ratio, fy=ratio)
    imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #Best for now
    blurred = cv2.GaussianBlur(imgray, (S_BLUR, S_BLUR), 0)
    edges = cv2.Canny(blurred, S_CANNY_TH1, S_CANNY_TH2)
    lines = cv2.HoughLines(edges, rho=1, theta=1 * np.pi / 180, threshold=S_HOUGH_TR)
    
    sect, keys = FindSectors(lines, edges)
    #print('Sections: ', sect)
    #print('Lines: ', len(lines))

    # First filter intersections then group
    # Find number of intersections for each line segment
    #AA = pairwise_distances(keys, metric=SectIntersection)
    #isects = sum(AA)

    # Find connected line segment to group
    #AA = pairwise_distances(keys[isects < S_INTERSEC], metric=SectIntersectionAll)
    #AA = (1 - AA) * 2
    #db = DBSCAN(eps=1, min_samples=2, metric="precomputed").fit(AA)
    #Z = db.labels_

    # Size of array for keys for filtered segments. Number of ungroupped (-1) plus 
    # number of groups
    #ungrouped, groups = np.sum(Z == -1), np.amax(Z) + 1
    
    #filt_seg = np.zeros((ungrouped + groups, 5))
    #filt_seg[0:ungrouped, :] = keys[isects < S_INTERSEC][Z == -1]
    
    #for i in range(0, np.amax(Z)+1):  
    #    filt_seg[ungrouped+i, :] = MergeSegments(keys[isects < S_INTERSEC][Z == i])

    # First group then filter intersections
    # Find connected line segment to group
    AA = pairwise_distances(keys, metric=SectIntersectionAll)
    AA = (1 - AA) * 2
    db = DBSCAN(eps=1.1, min_samples=2, metric="precomputed").fit(AA)
    Z = db.labels_

    # Size of array for keys for filtered segments. Number of ungroupped (-1) plus 
    # number of groups
    ungrouped, groups = np.sum(Z == -1), np.amax(Z) + 1
    
    filt_seg = np.zeros((ungrouped + groups, 5))
    filt_seg[0:ungrouped, :] = keys[Z == -1]
    
    for i in range(0, np.amax(Z)+1):  
        filt_seg[ungrouped+i, :] = MergeSegments(keys[Z == i])

    # Find number of intersections for each line segment using asymmetric distance
    AA = pair_distance(filt_seg, metric=SectIntersection)
    isects = sum(AA)

    # Find number of intersections for each line segment using symmetric distance
    a = filt_seg[isects < S_INTERSEC]
    AA = pairwise_distances(a, metric=SectIntersection)
    isects = sum(AA)

    return a[isects < S_INTERSEC], edges, blurred

    h, w = edges.shape[0:2]

    # Filter out np.pi * 0.15 < theta < np.pi * 0.85
    lines = lines[(np.pi * 0.15 > lines[:, 0, 1]) | (lines[:, 0, 1] > np.pi * 0.85)]

    # Get the x coordinate of the line at a middle horizontal line of the image
    # rho / np.cos(theta)) - 0.5 * h * np.tan(theta)
    order = lines[:, 0, 0] / np.cos(lines[:, 0, 1]) - 0.5 * h * np.tan(lines[:, 0, 1])

    # Sort by x coordinate on a middle horizontal line
    lines = lines[order.argsort()]

    contours = []
    first = True
    for l in lines:
        rho, theta = l[0]

        x1, y1 = int(rho / np.cos(theta)), 0
        x2, y2 = int(x1 - h * np.tan(theta)), h - 1

        rX = int((x1 + x2) / 2)

        if first:
            p_x1, p_y1, p_x2, p_y2, p_rX = x1, y1, x2, y2, rX
            first = False
        elif rX - p_rX > 10:
            contours.append([[[p_x1 / w, p_y1 / h], [x1 / w, y1 / h], [x2 / w, y2 / h], [p_x2 / w, p_y2 / h]], np.float(theta)])
            p_x1, p_y1, p_x2, p_y2, p_rX = x1, y1, x2, y2, rX

    return contours

def FindSectors(lines, img):
    h, w = img.shape[0:2]

    sections = []
    sec_keys = []
    for l in lines:
        rho, theta = l[0]
        
        # Process separetely lines near horizontal and vertical 
        # to avoid division to near 0
        if np.pi * 0.25 < theta < np.pi * 0.75:
            x1, y1 = 0, int(rho / np.sin(theta))
            x2, y2 = w - 1, int(y1 - w * np.tan(0.5 * np.pi - theta))
        else:
            x1, y1 = int(rho / np.cos(theta)), 0
            x2, y2 = int(x1 - h * np.tan(theta)), h - 1
        
        # find intercestion between line and image borders
        res, (x1, y1), (x2, y2) = cv2.clipLine((0, 0, w-2, h-2), (x1, y1), (x2, y2))
        
        if np.pi * 0.25 < theta < np.pi * 0.75:
            pts = np.zeros(((x2-x1+1)*S_WIDTH, 2), dtype=int)
            x = np.arange(x1, x2+1)
            y = np.uint(np.around((x - x1) * (y2 - y1) / (x2 - x1) + y1))
            pts[:,0] = np.repeat(x, min(S_WIDTH, 5))
            if S_WIDTH == 1:
                pts[:,1] = y
            elif S_WIDTH == 2:
                pts[:,1] = np.reshape([y, y+1], -1, order='F')
            elif S_WIDTH == 3:
                pts[:,1] = np.reshape([y-1, y, y+1], -1, order='F')
            elif S_WIDTH == 4:
                pts[:,1] = np.reshape([y-1, y, y+1, y+2], -1, order='F')
            elif S_WIDTH >= 5:
                pts[:,1] = np.reshape([y-2, y-1, y, y+1, y+2], -1, order='F')
        else:
            pts = np.zeros(((y2-y1+1)*S_WIDTH, 2), dtype=int)
            y = np.arange(y1, y2+1)
            x = np.uint(np.around((y - y1) * (x2 - x1) / (y2 - y1) + x1))
            pts[:,1] = np.repeat(y, min(S_WIDTH,5))
            if S_WIDTH == 1:
                pts[:,0] = x
            elif S_WIDTH == 2:
                pts[:,0] = np.reshape([x, x+1], -1, order='F')
            elif S_WIDTH == 3:
                pts[:,0] = np.reshape([x-1, x, x+1], -1, order='F')
            elif S_WIDTH == 4:
                pts[:,0] = np.reshape([x-1, x, x+1, x+2], -1, order='F')
            elif S_WIDTH >= 5:
                pts[:,0] = np.reshape([x-2, x-1, x, x+1, x+2], -1, order='F')
            
        # Exclude ponts outside the image
        pts = pts[(pts[:,0]>=0) & (pts[:,0]<w) & (pts[:,1]>=0) & (pts[:,1]<h) ]
        line = img[pts[:,1], pts[:,0]]
        #print('Line shape:', line.shape)

        dots = np.zeros((np.sum(line>0), 2))
        dots[:,0] = pts[:,0][line>0]
        dots[:,1] = pts[:,1][line>0]
        #print('Dots shape: ', dots.shape)
        
        clustering = DBSCAN(eps=DBSCAN_EPS, min_samples=DBSCAN_MIN_SAM).fit(dots)
        #print(clustering.labels_)

        for c in range(0, np.amax(clustering.labels_)+1):
            start_i = np.amin(np.where(clustering.labels_ == c))
            end_i = np.amax(np.where(clustering.labels_ == c))
            weight = np.sum(clustering.labels_ == c)
            pt_start = dots[start_i]
            pt_end = dots[end_i]
            sec_keys.append(SectToCartesian(w, h, pt_start, pt_end, theta, rho, weight))
            sections.append([[pt_start[0]/w, pt_start[1]/h], [pt_end[0]/w, pt_end[1]/h]])
            #print('Line: ', pt_start, pt_end)
    
    sections = np.array(sections)
    sec_keys = np.array(sec_keys)

    return sections[sec_keys[:,3]>S_MINLEN], sec_keys[sec_keys[:,3]>S_MINLEN]


def SectToCartesian(w, h, pt_start, pt_end, theta, rho, weight):
    if theta <= np.pi * 0.25:
        eps = (pt_start[1] + pt_end[1])/2/np.cos(theta) - rho * np.tan(theta)
    elif theta <= np.pi * 0.625:
        eps = rho / np.tan(theta) - (pt_start[0] + pt_end[0])/2/np.sin(theta)
    else:
        eps = rho * np.tan(theta) - (pt_start[1] + pt_end[1])/2/np.cos(theta)
    
    key_theta = theta
    
    if key_theta > np.pi * 0.625:
        key_theta = np.pi - theta

    # Length of the section
    delta = np.sqrt((pt_end[0] - pt_start[0])**2 + (pt_end[1] - pt_start[1])**2)
    return [theta, rho, eps, delta, weight]


def sign(a):
    b = np.sign(a)
    if isinstance(b, (list, tuple, np.ndarray)):
        b[b==0] = 1
    elif b==0:
        b = 1
    return b


def FindClusters(sections, trace=False):
    
    keys = np.zeros((sections.shape[0], 4))
    keys[:,0] = sections[:,0]
    keys[:,0][keys[:,0] > 0.625 * np.pi] = keys[:,0][keys[:,0] > 0.625 * np.pi] - np.pi
    keys[:,0] = S_K_THETA * keys[:,0]
    keys[:,1] = S_K_RHO * np.abs(sections[:,1])
    keys[:,2] = S_K_EPS * sections[:,2]
    keys[:,3] = S_K_DEL * sections[:,3]

    Z = ward(pdist(keys))
    Z = fcluster(Z, 600, criterion='distance')
    
    clusters = np.unique(Z)
    boxes = []
    for c in clusters:
        a = sections[Z==c]
        thetas = a[:,0]
        thetas[thetas > 0.625 * np.pi] = thetas[thetas > 0.625 * np.pi] - np.pi
        theta = np.average(thetas)

        rho_min, rho_max = np.amin(a[:,1] * sign(theta) * sign(thetas)), np.amax(a[:,1] * sign(theta) * sign(thetas))
        eps = np.average(a[:,2])
        #delta = np.average(a[:,3])
        delta = np.amax(a[:,3])

        rho_min = rho_min - (rho_max - rho_min) * S_RHO_EXT
        rho_max = rho_max + (rho_max - rho_min) * S_RHO_EXT
        
        if trace:
            print(a)
            print(c, ':', theta, [rho_min, rho_max], eps, delta)

        x1, y1 = FromCartesian(theta, rho_min, eps - delta/2*(1 + S_DEL_EXT))
        x2, y2 = FromCartesian(theta, rho_min, eps + delta/2*(1 + S_DEL_EXT))
        x3, y3 = FromCartesian(theta, rho_max, eps + delta/2*(1 + S_DEL_EXT))
        x4, y4 = FromCartesian(theta, rho_max, eps - delta/2*(1 + S_DEL_EXT))


        box = np.int32(np.array([[x1, y1], [x2, y2], [x3,y3], [x4, y4]]))
        boxes.append(box)

    return Z, np.array(boxes)


# As we do not want to delete longer segments with high number of intersections
# we have ASYMMETRIC distance matrix which is not supported by standard
def pair_distance(a, metric):
    dist = np.zeros((len(a), len(a)))
    for idx, x in enumerate(a):
        for jdx, xx in enumerate(a):
            dist[idx, jdx] = metric(xx, x)
    return dist


# Return 1 if line segments has an intersection, 0 otherwise
def SectIntersection(s1, s2):
    return SectIntersectionFiltered(s1, s2, S_K_ANGLE, np.pi, True)

# Return 1 if line segments has an intersection, 0 otherwise
def SectIntersectionAll(s1, s2):
    return SectIntersectionFiltered(s1, s2, 0, S_K_ANGLE, False)

# Return 1 if line segments has an intersection, 0 otherwise
def SectIntersectionFiltered(s1, s2, angle_min, angle_max, oneway):
    t1, r1, e1, d1 = s1[0:4]
    t2, r2, e2, d2 = s2[0:4]
    
    if t1 == t2 and r1 != r2:
        # No intersections for parallel lines
        intersect = 0
    elif t1 == t2 and r1 == r2:
        if abs(e1 - e2) < abs(d1+d2)/2:
            intersect = 1
        else:
            intersect = 0
    else:
        x1 = r2 / np.tan(t2 - t1) - r1 / np.sin(t2 - t1) 
        x2 = r1 / np.tan(t1 - t2) - r2 / np.sin(t1 - t2) 

        if t2 > np.pi * 0.625:
            x1 = -x1
        
        if t1 > np.pi * 0.625:
            x2 = -x2

        if e2 - d2/2 < x1 < e2 + d2/2 and e1 - d1/2 < x2 < e1 + d1/2:
            intersect = 1
        else:
            intersect = 0

    # if lines are almost parallel ignore intersections 
    if abs(t1 - t2) < angle_min:
        intersect = 0

    if angle_max < abs(t1 - t2) < np.pi - angle_max:
        intersect = 0

    # For filtering by number of intersections return 1 only for shorter segments 
    if oneway and d1 > d2:
        intersect = 0

    return intersect


def FromCartesian(theta, rho, eps):
    if theta < 0:
        theta = theta + np.pi
        
    if theta > np.pi * 0.625:
        eps = - eps

    x = - eps * np.sin(theta) + rho * np.cos(theta)
    y = eps * np.cos(theta) + rho * np.sin(theta)
    return int(x), int(y)

def MergeSegments(segm):
    t, r, e, d, w = np.copy(segm[:,0]), np.copy(segm[:,1]), np.copy(segm[:,2]), np.copy(segm[:,3]), np.copy(segm[:,4])
    
    t[t > np.pi * 0.625 ] = t[t > np.pi * 0.625 ] - np.pi
    r[t < 0] = -r[t < 0]
    #print('T:', t)
    #print('R:', r)
    
    theta = np.average(t)

    #print('TETHA:', theta)
    
    rho = np.average(r * np.cos(theta - t) + e * np.sin(theta - t))
    #print('RHO:', rho)

    eps = - np.sin(theta - t) * r + e * np.cos(theta - t)
    delta = abs(d * np.cos(theta - t))

    #print('EPS: ', eps)
    #print('DELTA: ', delta)

    mi = np.amin(eps - delta/2)
    ma = np.amax(eps + delta/2)
    
    if theta < 0:
        theta = theta + np.pi
        rho = -rho

    #print('RESULT: ', theta, rho, (ma+mi)/2, ma-mi)
    #return (theta, rho, (ma+mi)/2, ma-mi, np.sum(w)) # was perfect for 11/12
    return (theta, rho, (ma+mi)/2, ma-mi, np.amax(w))
```

```python
def BoxSegments(segm):
    t, r, e, d = np.copy(segm[:,0]), np.copy(segm[:,1]), np.copy(segm[:,2]), np.copy(segm[:,3])
    
    t[t > np.pi * 0.625 ] = t[t > np.pi * 0.625 ] - np.pi
    r[t < 0] = -r[t < 0]
    
    theta = np.average(t)

    rho_min = np.amin(r * np.cos(theta - t) + e * np.sin(theta - t))
    rho_max = np.amax(r * np.cos(theta - t) + e * np.sin(theta - t))
    #print('RHO:', rho_min, rho_max)

    eps = - np.sin(theta - t) * r + e * np.cos(theta - t)
    delta = abs(d * np.cos(theta - t))

    #print('EPS: ', eps)
    #print('DELTA: ', delta)

    mi = np.amin(eps - delta/2)
    ma = np.amax(eps + delta/2)

    # Extend along rho axis for the S_RHO_EXT of book height
    delta_avg = np.average(delta)
    rho_min = rho_min - delta_avg * S_RHO_EXT
    rho_max = rho_max + delta_avg * S_RHO_EXT
    
    if theta < 0:
        theta = theta + np.pi
        rho_min = -rho_min
        rho_max = -rho_max

    x1, y1 = FromCartesian(theta, rho_min, mi)
    x2, y2 = FromCartesian(theta, rho_min, ma)
    x3, y3 = FromCartesian(theta, rho_max, ma)
    x4, y4 = FromCartesian(theta, rho_max, mi)

    box = np.int32(np.array([[x1, y1], [x2, y2], [x3,y3], [x4, y4]]))

    return box


def FindClusters(sections, trace=False):
    
    # Find connectedline segment to group
    AA = pairwise_distances(sections, metric=StackedBooks)
    AA = (1 - AA) * 2
    db = DBSCAN(eps=1, min_samples=2, metric="precomputed").fit(AA)
    Z = db.labels_
    
    #keys = np.zeros((sections.shape[0], 4))
    #keys[:,0] = sections[:,0]
    #keys[:,0][keys[:,0] > 0.625 * np.pi] = keys[:,0][keys[:,0] > 0.625 * np.pi] - np.pi
    #keys[:,0] = S_K_THETA * keys[:,0]
    #keys[:,1] = S_K_RHO * np.abs(sections[:,1])
    #keys[:,2] = S_K_EPS * sections[:,2]
    #keys[:,3] = S_K_DEL * sections[:,3]

    #Z = ward(pdist(keys))
    #Z = fcluster(Z, 600, criterion='distance')
    
    if trace:
        print('Cluster vector: ',Z)

    clusters = np.unique(Z)
    boxes = []
    for c in range(0, max(Z)+1):
        box = BoxSegments(sections[Z==c])
        boxes.append(box)

    return Z, np.array(boxes)


# Return 1 if line segments has an intersection, 0 otherwise
def StackedBooks(s1, s2):
    t1, r1, e1, d1 = s1[0:4]
    t2, r2, e2, d2 = s2[0:4]
    
    if t2 > np.pi * 0.625:
        t2 = t2 - np.pi
        r2 = - r2
        
    if t1 > np.pi * 0.625:
        t1 = t1 - np.pi
        r1 = - r1

    # If not nearly parrallel it can't be a neighbouring books 
    if abs(t1 - t2) > S_PAR:
        #print('Angles: ', t1, t2)
        return 0
    
    # Check distance between segments if more than width of one book then disconnected 
    dist = abs(r1 / np.cos(t2 - t1) - r2 + np.tan(t2 - t1) * e2)
    if dist > max(d1, d2) * S_BOOKWIDTH:
        #print('Distance: ', dist, (r1 / np.cos(t2 - t1), r2, np.tan(t2 - t1) * e2))
        return 0
    
    e1p = - np.sin(t2 - t1) * r1 + e1 * np.cos(t2 - t1)
    d1p = abs(d1 * np.cos(t2 - t1))
    
    overlap = min(max((d2 + d1p)/2 - abs(e2 - e1p), 0), d2, d1p)/min(d2, d1p)
    if overlap < S_OVERLAP:
        #print('Overlap: ', overlap)
        return 0

    return 1


np.set_printoptions(precision=2, floatmode='fixed', suppress=True)
# Parameters of initial image preparation
# Size of the kernel for the blurring
#S_BLUR=7  # best
S_BLUR=5 # works for (9) not good for 13/14
# Canny edges threshold one and two
S_CANNY_TH1=5  # best
S_CANNY_TH2=200  # best
S_HOUGH_TR=80
# Parameters to identify segment of lines after HoughLine search:
# Minimum DBSCAN_MIN_SAM points in the DBSCAN_EPS distance in the line of width S_WIDTH 
# to consider it part of the line segment
S_WIDTH = 3
DBSCAN_EPS=20
#DBSCAN_MIN_SAM=20 # best
DBSCAN_MIN_SAM=30 # looks better than 20

# Parameters to filter noise segments after identifying segments 
# Keep only segments longer than S_MINLEN points and only segments with less 
# than S_INTERSEC intersections with other segments. Segments with angle more 
# than S_K_ANGLE are not connected
S_MINLEN = 70 # best
S_INTERSEC=3
S_K_ANGLE = 0.02 * np.pi # best
#S_K_ANGLE = 0.05 * np.pi # works for (9)

# Parameters to group segments into blocks of books stack together
S_PAR = 0.03 * np.pi
S_BOOKWIDTH = 0.25
S_OVERLAP = 0.85
S_TOOBIG = 2.1 # best
S_RHO_EXT = 0.3

IMAGEFILE = "shelves/12.jpg"
oimg = cv2.imread(IMAGEFILE)
ratio = 512/oimg.shape[1]
img = cv2.resize(oimg, dsize=(0, 0), fx=ratio, fy=ratio)

keys, edges, gray = get_contours(oimg)
print(edges.dtype)

print('Weight (min, avg, max): ', np.amin(keys[:,4]), np.average(keys[:,4]), np.amax(keys[:,4]))
#keys = keys[(keys[:,4] < 800) & (keys[:,4] > 100)] # Fantastic for img12, 13
keys = keys[(keys[:,4] < 250) & (keys[:,4] > 60)] # Fantastic for img12, 13

#plot_img(gray, show=True)

plot_img(edges, show=True)

t_img = np.copy(img)
color = (255, 0, 0)
for s in keys:
        theta, rho, eps, delta = s[0:4]
        x1, y1 = FromCartesian(theta, rho, eps - delta/2)
        x2, y2 = FromCartesian(theta, rho, eps + delta/2)
        cv2.line(t_img, (x1, y1), (x2, y2), color, 1)
plot_img(t_img, show=True)

#print('min, avg, median, max: ', np.amin(keys[:, 3]), np.average(keys[:, 3]), np.median(keys[:, 3]), np.amax(keys[:, 3]))
book_height = np.average(keys[:, 3])
no_big = keys[:, 3] < book_height * S_TOOBIG

Z, boxes = FindClusters(keys[no_big], trace=False)
clusters = np.unique(Z)

#print(clusters)

h, w = img.shape[0:2]

color = (0, 255, 0)
for s in keys[no_big][Z==-1]:
        theta, rho, eps, delta = s[0:4]
        x1, y1 = FromCartesian(theta, rho, eps - delta/2)
        x2, y2 = FromCartesian(theta, rho, eps + delta/2)
        cv2.line(img, (x1, y1), (x2, y2), color, 1)

for c in range(0, max(clusters)+1):  
    color = (np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255))
    for s in keys[no_big][Z==c]:
        theta, rho, eps, delta = s[0:4]
        x1, y1 = FromCartesian(theta, rho, eps - delta/2)
        x2, y2 = FromCartesian(theta, rho, eps + delta/2)
        cv2.line(img, (x1, y1), (x2, y2), color, 3)

    cv2.drawContours(img, [boxes[c]], 0, color, 3)
        
plot_img(img, show=True)
```

```python
S_K_THETA = 1
S_K_PARALLEL = 2
S_K_GAP = 100

def rho_dist(rho, rho1, alpha, eps):
    #print(rho1 / np.cos(alpha), - rho, - np.tan(alpha) * eps)
    return abs(rho1 / np.cos(alpha) - rho - np.tan(alpha) * eps)


def SectDistance(s1, s2):
    t1, t2 = s1[0], s2[0]
    
    if t1 > np.pi * 0.625:
        #e1 = - e1
        t1 = t1 - np.pi
        #r1 = - r1

    if t2 > np.pi * 0.625:
        #e2 = - e2
        t2 = t2 - np.pi
        #r2 = - r2

    theta = np.average((t1, t2))
    if theta < 0:
        theta = theta + np.pi

    #print('Average Theta: ', theta)

    # Restore original THETA
    t1, r1, e1 = s1[0:3]
    t2, r2, e2 = s2[0:3]
    
    # Choose rho as a average between projections of eps1 and eps2 to thetha
    rho = np.average((r1 * np.cos(theta - t1) + e1 * np.sin(theta - t1), r2 * np.cos(theta - t2) + e2 * np.sin(theta - t2)))
    #print('rho1: ', r1 * np.cos(theta - t1) + e1 * np.sin(theta - t1))
    #print('rho2: ', r2 * np.cos(theta - t2) + e2 * np.sin(theta - t2))
    rho_sign1 = sign(theta) * sign(t1) * rho
    rho_sign2 = sign(theta) * sign(t2) * rho
    
    #print('Average rho:', rho)

    eps1 = sign(t1 - t2) * sign(theta - t2) * (- np.sin(theta - t1) * r1 + e1 * np.cos(theta - t1))
    #print('EPS1: ', - np.sin(theta - t1) * r1,  e1 * np.cos(theta - t1))
    del1 = abs(s1[3] * np.cos(theta - t1))
    
    eps2 = sign(t2 - t1) * sign(theta - t1) * (- np.sin(theta - t2) * r2 + e2 * np.cos(theta - t2))
    #print('EPS2: ', - np.sin(theta - t2) * r2, e2 * np.cos(theta - t2))
    #print(e2, theta, t2)
    del2 = abs(s2[3] * np.cos(theta - t2))
    #print('EPS: ', (eps1, eps2))
    #print('Eps, del: ', eps1, del1, eps2, del2)

    if theta == t1: 
        mi = np.amin((eps1-del1/2, eps1+del1/2, eps2-del2/2, eps2+del2/2))
        ma = np.amax((eps1-del1/2, eps1+del1/2, eps2-del2/2, eps2+del2/2))
        dist = 2 * np.abs((rho - r1)*(ma - mi)) * S_K_PARALLEL

        proj = np.array([[theta, rho, eps1, del1], [theta, rho, eps2, del2]])
    else:
        x1 = rho / np.tan(theta - t1) - r1 / np.sin(theta - t1) 
        x2 = rho / np.tan(theta - t2) - r2 / np.sin(theta - t2)
        #print('X1: ',  rho / np.tan(theta - t1), - r1 / np.sin(theta - t1))
        #print('X2: ', rho / np.tan(theta - t2), - r2 / np.sin(theta - t2))

        if theta > np.pi * 0.625:
            x1, x2 = -x1, -x2

        #print('X1, X2: ', (x1, x2))

        mi, ma = eps1-del1/2, eps1+del1/2
        
        if mi > eps2 + del2/2:
            mi = eps2 + del2/2
            
        if ma < eps2 - del2/2:
            ma = eps2 - del2/2

        #print('X1 (min, max): ', (mi, ma))
        
        if mi <= x1 <= ma:
            d1 = 0.5 * (x1 - mi) * rho_dist(rho_sign1, r1, theta - t1, mi) + 0.5 * (ma - x1) * rho_dist(rho_sign1, r1, theta - t1, ma)
        else:
            d1 = 0.5 * (ma - mi) * (rho_dist(rho_sign1, r1, theta - t1, mi) + rho_dist(rho_sign1, r1, theta - t1, ma)) *  S_K_PARALLEL

        #print('D1: ', rho_dist(rho_sign1, r1, theta - t1, mi), rho_dist(rho_sign1, r1, theta - t1, ma))

        mi, ma = eps2-del2/2, eps2+del2/2
        
        if mi > eps2 + del2/2:
            mi = eps2 + del2/2
            
        if ma < eps2 - del2/2:
            ma = eps2 - del2/2

        #rint('X2 (min, max): ', (mi, ma))

        if mi <= x2 <= ma:
            d2 = 0.5 * (x2 - mi) * rho_dist(rho_sign2, r2, theta - t2, mi) + 0.5 * (ma - x2) * rho_dist(rho_sign2, r2, theta - t2, ma)
        else:
            d2 = 0.5 * (ma - mi) * (rho_dist(rho_sign2, r2, theta - t2, mi) + rho_dist(rho_sign2, r2, theta - t2, ma)) * S_K_PARALLEL
        
        #print('D2: ', rho_dist(rho_sign2, r2, theta - t2, mi), rho_dist(rho_sign2, r2, theta - t2, ma))
            
        dist = d1 + d2
        #dist = np.abs(((x1 - mi)**2 + (x1 - ma)**2 + (x2 - mi)**2 + (x2 - ma)**2) * np.tan(theta - s1[0]))

        if theta < 0:
            theta = theta + np.pi
            #eps1 = - eps1
            #eps2 = - eps2
        
        proj = np.array([[theta, rho, eps1, del1], [theta, rho, eps2, del2]])
        #print('Distance: ', d1, d2)

    if abs(eps1 - eps2) > (del1 + del2)/2:
        dist = dist +  S_K_GAP * (abs(eps1 - eps2) - (del1 + del2)/2)
    
    return dist
    #return dist, proj

S_PAR = 0.03 * np.pi
S_BOOKWIDTH = 0.2
S_OVERLAP = 0.7

sections = np.array([
    # Intercecting line segments with high angle
    [[0.25 * np.pi, 200, 88, 200], 
    [0.75 * np.pi, -10, 200, 200]],

    # Intercecting line segments with high angle
    [[0.25 * np.pi, 200, 0, 200], 
    [0.75 * np.pi, 0, 200, 200]],

    # Intercecting line segments 0.03
    [[0.83 * np.pi, 20, 270, 200], 
    [0.85 * np.pi, 3, 275, 200]],

    # Segments of the same line
    [[0.05 * np.pi, 200, 80, 100], 
    [0.05 * np.pi, 200, 200, 100]],

    # Parallel close to each other 0.15
    [[0.001 * np.pi, 198, 200, 200], 
    [0.998 * np.pi, -202, 200, 200]],

    # Connected lines 2.4
    [[0.11 * np.pi, 150, 10, 100], 
    [0.06 * np.pi, 138, 128, 100]],

    # Parallel with positive and negative RHO
    [[0.98 * np.pi, -2, 200, 200], 
    [0.98 * np.pi, 4, 200, 200]],

    # Parallel close to each other 0.15
    [[0.84 * np.pi, 10, 300, 200], 
    [0.84 * np.pi, 13, 270, 200]],

    # Parallel with shift 0.15
    [[0.84 * np.pi, 10, 300, 200], 
    [0.84 * np.pi, 50, 270, 200]],

    # Orthoganal lines 2.4
    [[0.1 * np.pi, 150, 10, 100], 
    [0.6 * np.pi, 300, -200, 100]],

    # Parallel 0.05
    [[0.84 * np.pi, 10, 310, 200], 
    [0.84 * np.pi, 50, 300, 200]],

    # Simmetric line segments 0.05
    [[0.83 * np.pi, 10, 310, 200], 
    [0.85 * np.pi, 10, 300, 200]],

    # Faraway lines 1.6
    [[0.2 * np.pi, 150, 10, 100], 
    [0.4 * np.pi, 300, -300, 100]]])

for s in sections:
    #dist, proj = SectDistance(s[0], s[1])
    dist = SectDistance(s[0], s[1])
    inters = SectIntersection(s[0], s[1])
    stacked = StackedBooks(s[0], s[1]) 
    print('Distance: ', dist)
    print('Intersected: ', inters)
    print('Stacked: ', stacked)

    img = np.zeros((384, 512), dtype=np.uint8)
    for l in s:  
        theta = l[0]
        rho = l[1]
        eps = l[2]
        delta = l[3]
    
        x1, y1 = FromCartesian(theta, rho, eps - delta/2)
        x2, y2 = FromCartesian(theta, rho, eps + delta/2)
    
        cv2.line(img, (x1, y1), (x2, y2), 255, 2)

    #print(proj)
    #for l in proj:  
    #    theta, rho, eps, delta = l[0:4]
     
    #    x1, y1 = FromCartesian(theta, rho, eps - delta/2)
    #    x2, y2 = FromCartesian(theta, rho, eps + delta/2)
    
    #    cv2.line(img, (x1, y1), (x2, y2), 128, 1)

    plot_img(img, show=True)
```

```python
def MergeSegments(segm):
    t, r, e, d = np.copy(segm[:,0]), np.copy(segm[:,1]), np.copy(segm[:,2]), np.copy(segm[:,3])
    
    t[t > np.pi * 0.625 ] = t[t > np.pi * 0.625 ] - np.pi
    r[t < 0] = -r[t < 0]
    print('T:', t)
    print('R:', r)
    
    theta = np.average(t)

    print('TETHA:', theta)
    
    rho = np.average(r * np.cos(theta - t) + e * np.sin(theta - t))
    print('RHO:', rho)

    eps = - np.sin(theta - t) * r + e * np.cos(theta - t)
    delta = abs(d * np.cos(theta - t))

    print('EPS: ', eps)
    print('DELTA: ', delta)

    mi = np.amin(eps - delta/2)
    ma = np.amax(eps + delta/2)
    
    if theta < 0:
        theta = theta + np.pi
        rho = -rho

    print('RESULT: ', theta, rho, (ma+mi)/2, ma-mi)
    return (theta, rho, (ma+mi)/2, ma-mi)


def FromCartesian(theta, rho, eps):
    if theta < 0:
        theta = theta + np.pi
        
    if theta > np.pi * 0.625:
        eps = - eps

    x = - eps * np.sin(theta) + rho * np.cos(theta)
    y = eps * np.cos(theta) + rho * np.sin(theta)
    return int(x), int(y)


S_K_THETA = 0.1

sections = np.array([
    # Parallel close to each other 0.15
    #[0.001 * np.pi, 198, 200, 200], 
    #[0.998 * np.pi, -202, 200, 200]
    #[0.02 * np.pi, 198, 100, 200], 
    #[0.98 * np.pi, -202, 100, 200]
    #[0.84 * np.pi, 10, 300, 200], 
    #[0.84 * np.pi, 13, 270, 200]
    # Two connected groups of segments with significantly different Theta
    [   2.23402143,   42.,  426.21403852,  111.82575732],
    [   1.57079637,  250., -148.00001093,  148.0135129 ],
    [   2.21656823,   49.,  402.72645188,  112.61438629],
    [   1.55334306,  252., -156.62585221,  174.02586015],
    [   2.28638124,   19.,  439.22959996,  101.31633629]
])

img = np.zeros((384, 512), dtype=np.uint8)
for l in sections:
        theta, rho, eps, delta = l[0:4]
    
        x1, y1 = FromCartesian(theta, rho, eps - delta/2)
        x2, y2 = FromCartesian(theta, rho, eps + delta/2)
    
        cv2.line(img, (x1, y1), (x2, y2), 255, 1)

theta, rho, eps, delta = MergeSegments(sections)
x1, y1 = FromCartesian(theta, rho, eps - delta/2)
x2, y2 = FromCartesian(theta, rho, eps + delta/2)
    
cv2.line(img, (x1, y1), (x2, y2), 128, 2)
        
plot_img(img, show=True)
```
